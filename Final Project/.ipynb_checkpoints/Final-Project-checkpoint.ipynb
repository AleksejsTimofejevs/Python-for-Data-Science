### Предобработка данных

**Заполнение пропусков**

```python
def Preprocessing(df):
    #Разбиение на дамми-переменные
    df['Helthcare_2'] = df['Helthcare_2'].asobject
    df = pd.get_dummies(df)
    
    #Заполнение пропусков в 'LifeSquare'
    na_predict = Lasso(alpha=0.03)
    
    LifeSquare_notna = df[(df['LifeSquare'].notna() == True)]
    LifeSquare_notna_target = LifeSquare_notna['LifeSquare']
    LifeSquare_notna_data = LifeSquare_notna.drop('LifeSquare', axis = 'columns')
    LifeSquare_notna_data = LifeSquare_notna_data.drop('Healthcare_1', axis = 'columns')
    try:
        LifeSquare_notna_data = LifeSquare_notna_data.drop('Price', axis = 'columns')
    except KeyError:
        pass
    
    LifeSquare_na = df[(df['LifeSquare'].notna() == False)]
    LifeSquare_na = LifeSquare_na.drop('LifeSquare', axis = 'columns')
    LifeSquare_na = LifeSquare_na.drop('Healthcare_1', axis = 'columns')
    try:
        LifeSquare_na = LifeSquare_na.drop('Price', axis = 'columns')
    except KeyError:
        pass
    
    na_predict.fit(LifeSquare_notna_data, LifeSquare_notna_target)
    
    na_predict_values = na_predict.predict(LifeSquare_na)
    
    df.loc[(df['LifeSquare'].isna()), 'LifeSquare'] = na_predict_values
    
    #Заполнение пропусков в 'Healthcare_1'
    na_predict = RandomForestRegressor(max_depth=6, n_estimators=300, random_state=42)
    
    df_healthcare_1_notna = df[(df['Healthcare_1'].notna() == True)]
    df_healthcare_1_notna_target = df_healthcare_1_notna['Healthcare_1']
    df_healthcare_1_notna_data = df_healthcare_1_notna.drop('Healthcare_1', axis = 'columns')
    try:
        df_healthcare_1_notna_data = df_healthcare_1_notna_data.drop('Price', axis = 'columns')
    except KeyError:
        pass
    
    df_healthcare_1_na = df[(df['Healthcare_1'].notna() == False)]
    df_healthcare_1_na = df_healthcare_1_na.drop('Healthcare_1', axis = 'columns')
    try:
        df_healthcare_1_na = df_healthcare_1_na.drop('Price', axis = 'columns')
    except KeyError:
        pass
    
    na_predict.fit(df_healthcare_1_notna_data, df_healthcare_1_notna_target)
    
    na_predict_values = na_predict.predict(df_healthcare_1_na)
    
    df.loc[(df['Healthcare_1'].isna()), 'Healthcare_1'] = na_predict_values
    return df
```

```python
df = Preprocessing(df)
```

### Построение моделей

**Разбиение данных на тренировочную и валидационную выборку**

```python
features = df.drop('Price', axis = 'columns').columns.tolist()
target = ['Price']
```

```python
X_train, X_valid,y_train, y_valid = train_test_split(df[features], df[target], test_size = 0.25, random_state = 20)
```

**Оптимальная модель Random Forest с кросс-валидацией**

```python
def R2(y_true,y_pred):    
     r2 = r2_score(y_true, y_pred)
     print('R2: %2.3f' % r2)
     return r2
    
def two_score(y_true,y_pred):
    score = R2(y_true,y_pred)
    return score

def two_scorer():
    return make_scorer(two_score, greater_is_better=True)


parameters = {
    'n_estimators': [200, 250, 350],
    'max_depth': np.arange(12, 18),
}

model = RandomForestRegressor()
clf = GridSearchCV(model, parameters, scoring=two_scorer())
```

```python
clf.fit(X_train, y_train)
```

```python
cv_results = pd.DataFrame(clf.cv_results_)

cv_results.columns
```

```python
param_columns = [
    column
    for column in cv_results.columns
    if column.startswith('param_')
]

score_columns = ['mean_test_score']

cv_results = (cv_results[param_columns + score_columns]
              .sort_values(by=score_columns, ascending=False))

cv_results.head(10)
```

```python
clf.best_params_
```

```python
best_model_params = clf.best_params_
```

```python
best_model = RandomForestRegressor(max_depth=best_model_params['max_depth'], n_estimators=best_model_params['n_estimators'])
```

```python
best_model.fit(X_train, y_train)
```

```python
y_pred = best_model.predict(X_valid)

r2_score(y_valid, y_pred)
```

**Протестируем модель на всех данных**

```python
all_pred = best_model.predict(df[features])
```

```python
r2_score(all_pred, df[target])
```
